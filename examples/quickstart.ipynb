## 1. 最小仿真 — 3行代码
创建一个地幔对流模拟并运行。

```python
from core.stokes_solver import StokesConfig, PicardStokesSolver

config = StokesConfig(
    nx=32, ny=32,          # 32×32 网格
    rayleigh=1e5,           # Rayleigh数
    max_picard_iterations=20,
    max_time_steps=100,
    dt=1e-3,
)

solver = PicardStokesSolver(config)
history = solver.run(n_steps=100)  # 带进度条!
```

## 2. 可视化结果
`plot()` 自动生成温度场、粘度场和速度场。

```python
solver.plot(field='all', save=False, show=True)
```

## 3. Nusselt数随时间变化

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(history['time'], history['nusselt'], 'b-', linewidth=2)
ax.axhline(y=10.53, color='r', linestyle='--', label='Blankenbach ref (Ra=10⁵)')
ax.set_xlabel('Time'); ax.set_ylabel('Nusselt Number')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

## 4. 保存结果

```python
solver.save('./my_convection_result')  # 生成 .npz + .vtk
```
